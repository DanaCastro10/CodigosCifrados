```python
def obtener_texto_plano_rc4():
    """Pide al usuario el texto plano y lo codifica a bytes para RC4."""
    while True:
        texto_plano = input("Ingrese el texto plano para cifrar (sin caracteres especiales): ")
        if texto_plano:
            return texto_plano.encode('utf-8')
        else:
            print("El texto plano no puede estar vacío. Inténtelo de nuevo.")
```

```python
def obtener_clave_rc4():
    """Pide al usuario una clave y la convierte a bytes."""
    while True:
        clave = input("Ingrese la clave para el cifrado RC4 (sin caracteres especiales): ")
        if clave:
            return clave.encode('utf-8')
        else:
            print("La clave no puede estar vacía. Inténtelo de nuevo.")
```

```python
def bytes_a_binario(data_bytes):
    """Convierte bytes a una cadena de bits (binario como texto)."""
    return ''.join(f'{byte:08b}' for byte in data_bytes)
```

```python
if __name__ == "__main__":
    # 1. Obtener entrada del usuario
    texto_en_bytes = obtener_texto_plano_rc4()
    clave_en_bytes = obtener_clave_rc4()

    # 2. Convertir el texto a binario
    binario_texto = bytes_a_binario(texto_en_bytes)
    print("\nTexto plano en binario:", binario_texto)

    # 3. Inicialización del estado S con KSA (Key Scheduling Algorithm)
    S = list(range(256))
    j = 0
    for i in range(256):
        j = (j + S[i] + clave_en_bytes[i % len(clave_en_bytes)]) % 256
        S[i], S[j] = S[j], S[i]  # intercambio

    # 4. PRGA (Pseudo Random Generation Algorithm)
    i = j = 0
    flujo_clave_bits = ''
    for _ in range(len(binario_texto)):
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]  # intercambio
        K = S[(S[i] + S[j]) % 256]
        flujo_clave_bits += f'{K:08b}'  # convertir a 8 bits

    # Ajustamos longitud para que coincida con el binario del texto
    flujo_clave_bits = flujo_clave_bits[:len(binario_texto)]

    # 5. Aplicamos XOR bit a bit
    texto_cifrado_binario = ''
    for bit_texto, bit_clave in zip(binario_texto, flujo_clave_bits):
        xor_bit = str(int(bit_texto) ^ int(bit_clave))  # XOR
        texto_cifrado_binario += xor_bit

    # 6. Convertimos el texto cifrado (en binario) a bytes
    texto_cifrado_bytes = bytes(
        int(texto_cifrado_binario[i:i+8], 2)
        for i in range(0, len(texto_cifrado_binario), 8)
    )

    # 7. Imprimimos los resultados
    print("\nTexto cifrado en bytes:", texto_cifrado_bytes)
    print("Texto cifrado en binario:", texto_cifrado_binario)
```
